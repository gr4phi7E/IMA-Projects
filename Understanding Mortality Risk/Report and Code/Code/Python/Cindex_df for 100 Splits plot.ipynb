```python
import pandas as pd
data1 = pd.read_csv("D:\\Final_Year_Project\\c_indices_Classical.csv")
data2 = pd.read_csv("D:\\Final_Year_Project\\c_index_LSTMdeepsurv.csv")
data3 = pd.read_csv("D:\\Final_Year_Project\\c_index_PCdeepsurv.csv")
```

```python
data2 = data2.rename(columns={'c_index': 'c_index_bilstmdeepsurv'})
```

```python
data2 = data2.drop('split', axis=1)
```

```python
data3 = data3.rename(columns={'c_index': 'c_index_deepsurv'})
```

```python
data3 = data3.drop('split', axis=1)
```

```python
df = pd.concat([data1, data3, data2], axis=1)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')

plt.figure(figsize=(14, 6))

for col in df.columns:
    plt.plot(df.index, df[col], label=col)

plt.xlabel('Split Index')
plt.ylabel('C-index')
plt.title('Comparative C-index across 100 Splits for Different Models')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
